## Investigation Overview

This project pretends to show based on the data of a bike service in San Francisco, the distribution of the rides and see how people is moving by bike within the cities covered by the bike service.

## Dataset Overview

The investigated dataset is the data provided by Ford GoBike from the year 2017 contained in a dataframe organized by biketrips. Each biketrip is described with information about the duration of the trip, and the start and end time and place. They also provide anonymized information about the users like, gender, birth year and type of user. It contains more than 500.000 trips and some of them were removed due to missing data. Finally the investigation was made in the remaining 453159 rows.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('2017-fordgobike-tripdata.csv')
```

```python
# wrangling data
df['start_time'] = pd.to_datetime(df.start_time)
df['end_time'] = pd.to_datetime(df.end_time)
df.start_station_id = df.start_station_id.astype(str)
df.end_station_id = df.end_station_id.astype(str)
df.bike_id = df.bike_id.astype(str)
df.dropna(inplace=True)
df.member_birth_year = df.member_birth_year.astype(int)
df['route'] = df['start_station_id'] + '-' + df['end_station_id']
```

## Distribution of trips during the day

The first thing that can be interesting is about the use that this bike system has. And in order to see that, a visualization per hour of the day is shown below. Where is possible to identify the peak times of a normal day. As expected at 8 AM and at 17 PM are the most crowded hours. It might be because people is using the bikes for quotidiane activities like go to work or school for example.

```python
ax = df.start_time.groupby(df["start_time"].dt.hour).count().plot(kind="bar",
                                                                  title='Bike trips per hour of the day');
```

## Distribution of trips during the week

One more step to understand how is people using the bikes in a regular way is to see the distribution of the bike trips during the days of the week. In this case is visible that in the weekends the use of the system is lower than during the working days, also during the working days the level of use of the bikes seems to be more or less uniform.

```python
df.start_time.groupby(df["start_time"].dt.weekday).count().plot(kind="bar",
                                                              title='Bike trips per day of the week');
```

## Distribution of rides per city

Seeing that the position of the stations was available in the dataset, it is interesting to see that this service covers three cities and that those three cities are San Francisco, San Jose and Oakland.
Looking at the latitude and longitude of the starting station for every trip is possible to identify clearly the three zones where the bikes are available.
Those zones were used to separate the database in three parts in order to visualize closely how the rides flow through the different cities.
In the next graphics we can see in green the stations where the rides starts and in red the arriving stations. There we can identify in every city an exchange of predominance in color between the morning and the afternoon. This we can understad like people going somewhere in the morning and going back in the afternoon.

```python
# create zones
sf_max_long = -122.35
oa_max_long = -122
start_zone = []
end_zone = []
for k, row in df.iterrows():
    if row['start_station_longitude'] < sf_max_long:
        start_zone.append('SF')
    elif row['start_station_longitude'] < oa_max_long:
        start_zone.append('OA')
    else:
        start_zone.append('SJ')
    if row['end_station_longitude'] < sf_max_long:
        end_zone.append('SF')
    elif row['end_station_longitude'] < oa_max_long:
        end_zone.append('OA')
    else:
        end_zone.append('SJ')
df['start_zone'] = start_zone
df['end_zone'] = end_zone
df.head()

# create cities
intercities = df[df['start_zone'] != df['end_zone']]
local = df[df['start_zone'] == df['end_zone']]
sanFrancisco = local[local['start_zone'] == 'SF']
oakland = local[local['start_zone'] == 'OA']
sanJose = local[local['start_zone'] == 'SJ']

# separate the city database in morning and afternoon
morning = sanFrancisco[sanFrancisco["start_time"].dt.hour < 12]
afternoon = sanFrancisco[sanFrancisco["start_time"].dt.hour >= 12]

# plotting morning
samples = np.random.choice(morning.index, 1000, replace = False)
rides_sample = morning.loc[samples,:]

ig, ax = plt.subplots(nrows = 2, figsize = [7,10])
routes_count = rides_sample.groupby('route')['duration_sec'].count()
routes_x1 = rides_sample.groupby('route')['start_station_longitude'].mean()
routes_x2 = rides_sample.groupby('route')['end_station_longitude'].mean()
routes_y1 = rides_sample.groupby('route')['start_station_latitude'].mean()
routes_y2 = rides_sample.groupby('route')['end_station_latitude'].mean()
ax[0].set_title('Morning Rides in San Francisco')
ax[0].set_xlabel('longitude')
ax[0].set_ylabel('latitude')
for i in range(len(routes_count)):
    y = [routes_y1[i], routes_y2[i]]
    x = [routes_x1[i], routes_x2[i]]
    size= routes_count[i]
    ax[0].plot(x, y, color='b', alpha=0.3, linewidth=5 * size / max(routes_count))
    ax[0].plot(x[0], y[0], color='g', alpha=0.2, marker='o', markersize=5 * size)
    ax[0].plot(x[1], y[1], color='r', alpha=0.2, marker='o', markersize=5 * size)

#plotting afternoon
samples = np.random.choice(afternoon.index, 1000, replace = False)
rides_sample = afternoon.loc[samples,:]

routes_count = rides_sample.groupby('route')['duration_sec'].count()
routes_x1 = rides_sample.groupby('route')['start_station_longitude'].mean()
routes_x2 = rides_sample.groupby('route')['end_station_longitude'].mean()
routes_y1 = rides_sample.groupby('route')['start_station_latitude'].mean()
routes_y2 = rides_sample.groupby('route')['end_station_latitude'].mean()
ax[1].set_title('Afternoon Rides in San Francisco')
ax[1].set_xlabel('longitude')
ax[1].set_ylabel('latitude')
for i in range(len(routes_count)):
    y = [routes_y1[i], routes_y2[i]]
    x = [routes_x1[i], routes_x2[i]]
    size= routes_count[i]
    ax[1].plot(x, y, color='b', alpha=0.3, linewidth=5 * size / max(routes_count))
    ax[1].plot(x[0], y[0], color='g', alpha=0.2, marker='o', markersize=5 * size)
    ax[1].plot(x[1], y[1], color='r', alpha=0.2, marker='o', markersize=5 * size)
```
