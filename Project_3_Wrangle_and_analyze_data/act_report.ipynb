The question of the title might seem hard to answer but let's get an approach looking at the database collected through tweeter and the public opinion.
We were provided with the tweets database of WeRateDogs where people were posting pictures about thir dogs and rating them with numbers out of 10. And after wrangling the database we wanted to check if the rating provided by the people twitting was actually supported by the community.
So this is what we've got.

```python
import pandas as pd
df = pd.read_csv('twitter_archive_master.csv')
```

```python
df_filtered = df.query('rating_out_of_10 < 20')
df_filtered.plot(x='rating_out_of_10', y='favorite_count', kind='scatter')
df_filtered.plot(x='rating_out_of_10', y='retweet_count', kind='scatter');
```

So based on the graphics we can get, it is possible to see that low rated dogs tend to have less retweet amounts and also les favorite amounts. At the same time, the more are the dogs rated, the higher values of retweet count and favorite count start to appear which can not be taken as a proof of honesty from dog owners but at least can give us a feeling that when somebody says that his pet is very good, even 13 or 14 out of 10, it might be true and it might be supported by huge communities of people like tens of thousands!
